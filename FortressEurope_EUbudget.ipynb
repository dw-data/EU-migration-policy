Unit: Million EUR

```python
EU_security_budget = pd.read_csv("data/EU-Security-Budget.csv")
EU_security_budget
```

```python
EU_security_budget.info()
```

```python
fig, ax = plt.subplots(figsize = (10,10))

EU_security_budget.plot(kind = "line", x="year", y="asylum_migration_integration", ax = ax, c="green")
EU_security_budget.plot(kind = "line", x="year", y="internal_security_funds", ax = ax, c="red")
# EU_security_budget.plot(kind = "line", x="year", y="decentralized_agencies", ax = ax, c="blue")

ax.set_ylim(0,2000)
ax.set_ylabel("Million EUR")
ax.set_xtickslabels = ["2014", "2015", "2016", "2017", "2018"]

# plt.savefig("VisualDrafts/EU-budget.pdf", transparent = True)
```

```python
EU_security_budget["share_asylum_in_security"]=EU_security_budget["asylum_migration_integration"]/EU_security_budget["security_citizenship"]*100
EU_security_budget["share_intsecurity_in_security"]=EU_security_budget["internal_security_funds"]/EU_security_budget["security_citizenship"]*100
EU_security_budget["share_decentagencies_in_security"]=EU_security_budget["decentralized_agencies"]/EU_security_budget["security_citizenship"]*100
```
